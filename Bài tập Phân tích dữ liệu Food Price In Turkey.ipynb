# Thang đo tương ứng của các thuộc tính được lưu trữ

1. Place, ProductName, UmName: có kiểu object --> dữ liệu định danh Norminal
2. ProductId, UmId, Month, Year: có kiểu số nguyên(int64) mô tả mã số sả phẩm mã số của đơn vị đo lường,
tháng và năm quan sát. \
Những thuộc tính này dùng để định danh, phân biệt ==> dữ liệu định danh Norminal
3. Price: mô tả thông tin đếm được, làm được các phép cộng, trừ, nhân, chia, có số'0' tuyệt đối ==> thang đo Ratio

# Giá trung bình của từng loại thực phẩm là bao nhiêu

```python
df_by_Productname=df.groupby('ProductName')
mean_price=df_by_Productname['Price'].aggregate(np.mean)
print (f"Giá trung bình của từng loại thực phẩm : {mean_price}")
```
